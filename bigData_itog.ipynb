```python
#Этап1.С сайта Банка России скачайте данные о курсе рубля по отношению к доллару США за последние 5 лет.
#Данные можно скачать в формате Excel или с помощью API Банка России. 
#Преобразуйте ежедневные показатели в ежемесячные.
rate = pd.read_excel('C:\\Users\\koves\\Desktop\\Magistratura\\Python\\RC_F01_12_2018_T01_12_2023.xlsx')
rate_data = rate[['data','curs']]
rate_data.head()
rate_data.set_index('data', inplace=True)
rate_data.index = pd.to_datetime(rate_data.index)
rate_data.resample('1M').mean()
rate_final = rate_data.reset_index()
```

```python
#Этап 2
#С сайта СтатБюро скачайте ежемесячные данные об индексе потребительских цен в России за последние 5 лет.
cpi=pd.read_excel('C:\\Users\\koves\\Desktop\\Magistratura\\Python\\cpi.xlsx')
```

```python
#Этап 3
#С помощью Python постройте графики динамики курса рубля и инфляции за последние 5 лет.
plt.figure(figsize=(16,10), dpi=80)
plt.plot('data', 'curs', data=rate_final, color='tab:red')
plt.plot('date','cpi',data=cpi, color='tab:blue')
plt.title('График динамики курса рубля и инфляции за последние 5 лет')
```
