```python
import pandas as pd
movies = pd.read_csv('C:\\Users\\koves\\Desktop\\Magistratura\\Python\\movies.csv')
ratings = pd.read_csv('C:\\Users\\koves\\Desktop\\Magistratura\\Python\\ratings.csv')    
```

```python
def class_movies(data):
    if data['rating'] <= 2:
        rate = 'низкий рейтинг'
    elif data['rating'] <= 4:
        rate = 'cредний рейтинг'
    elif data['rating'] == 4.5 or data['rating'] == 5:
        rate = 'высокий рейтинг'
    else:
        rate = 'не хватает условий в задаче'
    return rate

ratings['class'] = ratings.apply(class_movies, axis=1)
ratings.head(100)
```
