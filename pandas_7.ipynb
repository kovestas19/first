```python
#Первый график. 
#Выбрал рост и год рождения баскетболистов. 
#По графику можно понять, что за 80 лет рост баскетболистов менялся не сильно
#Выбрал график рассеивания plot, потому что он лучше всего показывает разброс 
plt.plot(players.born,players.height)
```

```python
#Второй график
#Количество очков и позицию баскетболиста, чтобы увидеть взаимосвязь позиции игрока на поля и количество очков,
#которые он зарабатывает
#Использовал круговую диагрумму, т.к.она лучше всего подходит для данной выборки
#

stats_1 = stats[['Pos','PTS']]
pos_pts = stats_1.groupby('Pos').sum('PTS').sort_values(['PTS']).reset_index().head(10)
#pos_pts.plot(kind='pie',y ='PTS',x='Pos')

pos_pts.groupby(['Pos']).sum().plot(kind='pie', y='PTS')
plt.legend(bbox_to_anchor=(0.6, .5, 0.6, 0.2))
```

```python
# Третий график
#Коррелиация между ростом и весом игроков. Цель увидеть взаимосвязь между ростом и весом игроков
#Использовал sns т.к. он лучше подходит в данном случае

import seaborn as sns
sns.regplot(data=players, x='height', y='weight', color='red', marker='.')
plt.title('Корреляция между ростом и весом игроков')
plt.ylim(0, 200)
```

```python
m1 = movies[['release_date','popularity','budget']].groupby('release_date').head(1000)
#df['year'] = pd.DatetimeIndex(df['ArrivalDate']).year
m1['year']= pd.DatetimeIndex(m1['release_date']).year
m2 = m1[['year','popularity','budget']].groupby('year').sum()
m2['budget'] = m2['budget']/1000
m2['popularity'] = m2['popularity']*1000
m3 = m2.sort_values('year',ascending = False).head(40)
m3.plot(kind='bar', stacked=True)
```
