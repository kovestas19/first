```python
import numpy as np, matplotlib as mpl, matplotlib.pyplot as plt, pandas as pd
import seaborn as sns, math, os, warnings 
from xgboost import XGBClassifier
from sklearn.model_selection import train_test_split

warnings.filterwarnings('ignore')
```

```python
uplift = pd.read_csv('C:\\Users\\koves\\Desktop\\Magistratura\\Python\\uplift_synthetic_data_100trials.csv')
uplift.head(6)
```

```python
uplift['treatment_group_key'].unique()
```

```python
#Определяем классы клиентов:
#Control Non-Responders(CN) treatment_group_key = control; conversion = 0
#Control Responders(CR) treatment_group_key = control; conversion = 1
#Treatment Non-Responders(TN) treatment_group_key = treatment1; conversion = 0
#Treatment Non-Responders(TR) treatment_group_key = treatment1; conversion = 1
```

```python
cn = uplift.loc[((uplift['treatment_group_key']== 'control')&(uplift['conversion']== 0))]
cr = uplift.loc[((uplift['treatment_group_key']== 'control')&(uplift['conversion']== 1))]
tn = uplift.loc[((uplift['treatment_group_key']== 'treatment1')&(uplift['conversion']== 0))]
tr = uplift.loc[((uplift['treatment_group_key']== 'treatment1')&(uplift['conversion']== 1))]
x_cn,y_cn = cn[['control_conversion_prob']],cn[['treatment1_true_effect']]
x_cr,y_cr = cr[['control_conversion_prob']],cr['treatment1_true_effect']
x_tn,y_tn = tn[['treatment1_conversion_prob']],tn['treatment1_true_effect']
x_tr,y_tr = tr[['treatment1_conversion_prob']],tr['treatment1_true_effect']

x_cn_train,x_cn_test,y_cn_train,y_cn_test = train_test_split(x_cn,y_cn.index,test_size = 0.2)
x_cr_train,x_cr_test,y_cr_train,y_cr_test = train_test_split(x_cr,y_cr,test_size = 0.2)
x_tn_train,x_tn_test,y_tn_train,y_tn_test = train_test_split(x_tn,y_tn,test_size = 0.2)
x_tr_train,x_tr_test,y_tr_train,y_tr_test = train_test_split(x_tr,y_tr,test_size = 0.2)
```
