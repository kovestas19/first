1.В переменную years запишите список из всех годов с 1950-го по 2010-й.

```python
years = list(range(1950,2011))
print(years)
```

2.Напишите функцию production_year, которая каждой строке из названия фильма выставляет год выпуска (год выпуска берём из столбца title). Не все названия фильмов содержат год выпуска в одинаковом формате, поэтому используйте следующий алгоритм:

для каждой строки пройдите по всем годам списка years;
если номер года присутствует в названии фильма, то функция возвращает этот год как год выпуска;
если ни один из номеров года списка years не встретился в названии фильма, то возвращается 1900 год.

```python
def production_year(df,lst):
    result = 0
    string = df['title']
    for i,r in enumerate(lst):
        if str(r) in string:
            result = r
    if result == 0:
        return 1900
    else:
        return result
```

3.Запишите год выпуска фильма по алгоритму пункта 2 в новый столбец ‘year’.

```python
movies_stats['year'] = movies_stats.apply(production_year,lst = years, axis=1)    
movies_stats.head()
```

4.Посчитайте средний рейтинг всех фильмов для каждого значения столбца ‘year’ и отсортируйте результат по убыванию рейтинга.

```python
results = movies_stats.groupby('year').agg({'rating':'mean'}).reset_index()
results.sort_values(by='rating', ascending=False)
```
